```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
data = pd.read_csv('trycsv2.csv')

Iterations = np.asarray(data.iloc[0,1:])
StepsValue = np.asarray(data.iloc[1,1:])
TimeValue = np.asarray(data.iloc[4,1:])
RewardsValue = np.asarray(data.iloc[7,1:])

StepsPolicy = np.asarray(data.iloc[2,1:])
TimePolicy = np.asarray(data.iloc[5,1:])
RewardsPolicy = np.asarray(data.iloc[8,1:])

StepsQ = np.asarray(data.iloc[3,1:])
TimeQ = np.asarray(data.iloc[6,1:])
RewardsQ = np.asarray(data.iloc[9,1:])
```

```python
plt.figure(figsize=(10,5))
plt.plot(Iterations, StepsValue,label = 'Value Iteration')
plt.plot(Iterations, StepsPolicy, label = 'Policy Iteration')
plt.plot(Iterations, StepsQ, label = 'Q-Learning')
plt.xlabel('Iterations')
plt.ylabel('Number of actions')
plt.title('Number of actions required by the robot to reach terminal state')
plt.legend(loc="best")
plt.ylim(0,2000)

plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(Iterations, TimeValue,label = "Value Iterations")
plt.plot(Iterations, TimePolicy, label = "Policy Iterations")
plt.plot(Iterations, TimeQ, label = "Q learning")
plt.xlabel('Iterations')
plt.ylabel('Time(ms)')
plt.title('Time taken(in ms) per iteration')
plt.legend(loc="best")
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(Iterations, RewardsValue,label = "Value Iterations")
plt.plot(Iterations, RewardsPolicy, label = "Policy Iterations")
plt.plot(Iterations, RewardsQ, label = "Q learning")
plt.xlabel('Iterations')
plt.ylabel('Rewards')

plt.legend(loc="best")
plt.ylim(-800,100)
plt.show()
```
