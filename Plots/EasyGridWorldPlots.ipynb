```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
data = pd.read_csv('trycsv.csv')
data2 = pd.read_csv('selectepsilon.csv')

epsilon0_2 = np.asarray(data2.iloc[1,1:])
epsilon0_15 = np.asarray(data2.iloc[2,1:])
epsilon0_1 = np.asarray(data2.iloc[3,1:])

Iterations = np.asarray(data.iloc[0,1:])
StepsValue = np.asarray(data.iloc[1,1:])
TimeValue = np.asarray(data.iloc[4,1:])
RewardsValue = np.asarray(data.iloc[7,1:])

StepsPolicy = np.asarray(data.iloc[2,1:])
TimePolicy = np.asarray(data.iloc[5,1:])
RewardsPolicy = np.asarray(data.iloc[8,1:])

StepsQ = np.asarray(data.iloc[3,1:])
TimeQ = np.asarray(data.iloc[6,1:])
RewardsQ = np.asarray(data.iloc[9,1:])

```

```python
plt.figure(figsize=(10,5))
plt.plot(Iterations, epsilon0_1,label='epsilon=0.1')
plt.plot(Iterations, epsilon0_15,label = 'epsilon=0.15')
plt.plot(Iterations, epsilon0_2,label = 'epsilon=0.2')
plt.xlabel('Iterations')
plt.ylabel('No of actions')
plt.title('Choosing the optimal value of epsilon')
plt.legend(loc="best")
plt.ylim(0, 600)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(Iterations, StepsValue)
plt.plot(Iterations, StepsPolicy)
plt.plot(Iterations, StepsQ)
plt.xlabel('Iterations')
plt.ylabel('Number of actions')
plt.title('Number of actions required by the robot to reach terminal state')
plt.ylim(0, 600)

plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(Iterations, TimeValue,label = "Value Iterations")
plt.plot(Iterations, TimePolicy, label = "Policy Iterations")
plt.plot(Iterations, TimeQ, label = "Q learning")
plt.xlabel('Iterations')
plt.ylabel('Time(ms)')
plt.title('Time taken(in ms) per iteration')
plt.legend(loc="best")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(Iterations, RewardsValue,label = "Value Iterations")
plt.plot(Iterations, RewardsPolicy, label = "Policy Iterations")
plt.plot(Iterations, RewardsQ, label = "Q learning")
plt.xlabel('Iterations')
plt.ylabel('Rewards')
plt.title('Rewards following the current policy per iteration')
plt.legend(loc="best")
plt.ylim(-300,100)
plt.show()
```
